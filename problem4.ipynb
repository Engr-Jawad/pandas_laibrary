# medium levle problems in pandas

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "Ayesha", "David", "Hina", "Ahmed"],
    "Math": [78, 92, 65, 88, 90, 95, 72],
    "Physics": [85, 91, 70, 60, 89, 98, 75],
    "Attendance": [90, 96, 70, 85, 88, 99, 80],
    "Department": ["CS", "IT", "CS", "SE", "IT", "CS", "SE"]
}

df = pd.DataFrame(data)
# Show students who belong to "CS" department and math marks are less than 75
cs_Std = df.loc[(df["Department"] == "CS") & (df["Math"] < 75 ) ]
print(cs_Std)
print("----------------------------------------------")
# Show students who Attendance is less than 85  OR Physics marks are less than 72
filter_std = df.loc[(df["Attendance"] < 85 ) | (df["Physics"] <72)]
print("students who Attendance is less than 85  OR Physics marks are less than 72\n",filter_std)
print("----------------------------------------------")
# Show students who are NOT from "IT" department
not_it_std = df.loc[df["Department"] != "IT"]
print("student not from IT department : \n",not_it_std)
print("----------------------------------------------")
# Show students whose Math marks are between 70 and 90
math_70_90 = df.loc[(df["Math"] > 70 )& (df["Math"] < 90)]
print("student whoes marks between 70 and 90 \n",math_70_90)
print("----------------------------------------------")

# show name ,math,department
result = df.loc[df["Attendance"] > 85 ,["Name","Math","Department"]]
print(result)
print("----------------------------------------------")
# Sort only CS students by Math marks in descending order.
sort = df[df["Department"]== "CS"].sort_values(by="Math",ascending=False ) 
print(sort)

df
```
