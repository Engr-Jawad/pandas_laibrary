# medium levle problems in pandas

```python

import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Sara", "John", "Ayesha", "David", "Hina", "Ahmed"],
    "Math": [78, 92, 65, 88, 90, 95, 72],
    "Physics": [85, 91, 70, 60, 89, 98, 75],
    "Attendance": [90, 96, 70, 85, 88, 99, 80],
    "Department": ["CS", "IT", "CS", "SE", "IT", "CS", "SE"]
}

df = pd.DataFrame(data)
# Show students who belong to "CS" department and math marks are less than 75
cs_Std = df.loc[(df["Department"] == "CS") & (df["Math"] < 75 ) ]
print(cs_Std)
print("----------------------------------------------")
# Show students who Attendance is less than 85  OR Physics marks are less than 72
filter_std = df.loc[(df["Attendance"] < 85 ) | (df["Physics"] <72)]
print("students who Attendance is less than 85  OR Physics marks are less than 72\n",filter_std)
print("----------------------------------------------")
# Show students who are NOT from "IT" department
not_it_std = df.loc[df["Department"] != "IT"]
print("student not from IT department : \n",not_it_std)
print("----------------------------------------------")
# Show students whose Math marks are between 70 and 90
math_70_90 = df.loc[(df["Math"] > 70 )& (df["Math"] < 90)]
print("student whoes marks between 70 and 90 \n",math_70_90)
print("----------------------------------------------")

# show name ,math,department
result = df.loc[df["Attendance"] > 85 ,["Name","Math","Department"]]
print(result)
print("----------------------------------------------")
# Sort only CS students by Math marks in descending order.
sort = df[df["Department"]== "CS"].sort_values(by="Math",ascending=False ) 
print(sort)

df
```

Calculate the percentage contribution of each row to the overall sales.
Create a new column:

"Good" if Rating ≥ 4.0 and < 4.5
"Average" otherwise

```python

data = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Laptop",
                "Mouse", "Keyboard", "Monitor", "Laptop", "Mouse"],
    "Category": ["Electronics", "Accessories", "Accessories", "Electronics",
                 "Electronics", "Accessories", "Accessories", "Electronics",
                 "Electronics", "Accessories"],
    "Price": [80000, 1500, 3000, 25000, 82000, np.nan, 3200, 26000, 79000, 1800],
    "Quantity": [5, 20, 10, 4, 3, 15, np.nan, 6, 2, 18],
    "Rating": [4.5, 4.2, 4.0, np.nan, 4.7, 4.1, 3.9, 4.4, 4.8, 4.3]
}

df = pd.DataFrame(data)
# Display the first 5 rows.
df.head(5)
# Find the number of missing values in each column.
df.isnull().sum()
# Replace missing Price values with the mean Price.
mean = df["Price"].mean()
df["Price"]= df["Price"].fillna(mean)
# Replace missing Quantity values with the median Quantity.
median = df["Quantity"].median()
df["Quantity"] = df["Quantity"].fillna(median)
# Replace missing Rating values with the average Rating.
avg_rat = df["Rating"].mean()
df["Rating"] = df["Rating"].fillna(avg_rat)
# Create a new column Total_Sales = Price × Quantity.
df["total_Sales"] =df["Price"] * df["Quantity"]
# Find the product with the highest Total_Sales.
highest_sales = df.loc[df["total_Sales"].idxmax()]
print("highest_sales product : \n",highest_sales)
print("\n")
# Calculate the total sales for each Category.
category_sales = df.groupby("Category")["total_Sales"].sum()
print(category_sales)
print("\n")
# Find the average Rating for each Product.
average_rat = df.groupby("Product")["Rating"].mean()
print(average_rat)
print("\n\n")
# Sort the DataFrame by Total_Sales in descending order.
sort = df.sort_values(by="total_Sales",ascending=False)
print(sort)
# Display only rows where Rating > 4.3
filter = df.loc[np.where(df["Rating"] > 4.3)]
print(filter)
# Find the top 3 products by total sales.
top_3 = df.sort_values(by="total_Sales",ascending=False).head()
print(top_3)

df
```

# merging and concat problems

```python


# Sales data for Q1 and Q2
q1_sales = pd.DataFrame({
    'product': ['A', 'B', 'C'],
    'revenue': [1000, 1500, 800],
    'quantity': [10, 15, 8]
})

q2_sales = pd.DataFrame({
    'product': ['A', 'B', 'D'],
    'revenue': [1200, 1400, 900],
    'quantity': [12, 14, 9]
})

# Task: Concatenate Q1 and Q2 sales data
# Question: What happens to the indices when you concatenate?
# answer : indices are remain not changing 0 twice and 1 twice like thiss
# Expected output: All rows combined
pd.concat([q1_sales,q2_sales],ignore_index=True)
```

```python
# Different columns in each DataFrame
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'id': [4, 5],
    'age': [25, 30],
    'city': ['NYC', 'LA']
})

# Task 1: Concatenate with default behavior (columns don't match)
# Task 2: Concatenate and fill missing values with 0
pd.concat([df1,df2]).fillna(0)
```

```python
# Sales and product info
sales = pd.DataFrame({
    'sale_id': [101, 102, 103],
    'amount': [500, 600, 550]
})

product_info = pd.DataFrame({
    'product': ['Laptop', 'Mouse', 'Keyboard'],
    'category': ['Electronics', 'Accessories', 'Accessories']
})

# Task: Concatenate horizontally (axis=1)
# Question: What assumption does this make about row alignment?
pd.concat([sales,product_info],axis=1,ignore_index=True)
```
