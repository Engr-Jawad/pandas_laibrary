# Data structure in pandas

### series
<p> it is defined as a one dimensional aray that is capable of storing various data types

```python
list =[9,8,7,6,5,4,3]
var = pd.Series(list)
#  series is one dimensional pandas data structure which store value in one dimension
print("series  data structure : \n",var)
#  type 
print(type(var))
# inserting new element at index 5 
var[5]=20
print("insert new element at index 5  : ",var)
# converting data type from one to other form
print("converting data type frome int64 to int 32   :",pd.Int32Dtype)
```
