# Problem: Export Employee Data to CSV
You are given a dictionary containing employee records. Your task is to:

Create a pandas DataFrame from the data.

Save the DataFrame to a CSV file named employees.csv with the following requirements:

Exclude the index column from the CSV.

Use a semicolon (;) as the delimiter instead of comma.

Do not include the column headers in the output file.

Only save rows where Salary is greater than 50,000 (filter before saving).

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'HR'],
    'Salary': [55000, 72000, 48000, 91000, 46000],
    'Years_Exp': [3, 5, 2, 8, 1]
}
data_f = pd.DataFrame(data)
data_f.to_csv("employees.csv",index=False,sep=";",header=False)
data_f[data_f['Salary'] > 50000]
data_f
```

# student performance analysis

### 1. Create a new column called "Average"

Calculate the average marks of each student.

```python


data = {
    "Name": ["Ali", "Sara", "John", "Ayesha", "David", "Zain"],
    "Math": [78, 92, 85, 60, 88, 73],
    "Physics": [80, 95, 78, 65, 90, 70],
    "Chemistry": [75, 91, 82, 70, 85, 68]
}

df = pd.DataFrame(data)
df
df["average"] = df[["Math","Physics","Chemistry"]].mean(axis=1)
df
```

### 2. Create a new column called "Result"

If average marks are greater than or equal to 75 → "Pass"
Otherwise → "Fail"

```python
df["result"] = np.where(df["average"] >= 75 ,"pass","fail")
df
```

### Find the student with the highest average marks.

```python
df.loc[df["average"].idxmax()]
```

```python


np.random.seed(10)

data = {
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard", "Mouse"],
    "Price": np.random.randint(500, 5000, 6),
    "Quantity": np.random.randint(1, 15, 6),
    "Discount": np.random.randint(5, 30, 6)
}

df = pd.DataFrame(data)

print(df)
```

# Create a new column:

```python
df["total"] = df["Price"] * df["Quantity"]
df
```

```python
df["final_price"] = df["total"] - ((df["total"] * df["Discount"])/100)
df
```

# 3.

Product with highest Final_Price
Product with lowest Final_Price

```python
# product with highest final price
print(df.loc[df["final_price"].idxmax()])
print("\n\n")
# product with lowest final price
print(df.loc[df["final_price"].idxmin()])

```
