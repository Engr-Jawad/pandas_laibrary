# How to insert and deleete data in data frame
# syntax
variable.insert(location,"name",[value])

```python
import pandas as pd
```

```python
data = {"a":[1,2,3,4],"b":[4,3,2,2]}
dataframe = pd.DataFrame(data)
dataframe
# coloumns c are inserted 
dataframe.insert(2,"c",[9,8,7,6])
dataframe
# columns d are inserted
dataframe.insert(3,"d",["a","s","d","f"])
dataframe
```

# coying data from one columns to another

```python
# copying all the elements of column a to e
dataframe["e"] = dataframe["a"][:]
# copying elements from column b to columns f from index o to 3 but three are exclusive
dataframe["f"] = dataframe["b"][:3]
dataframe
```

# deleting data from dataframe

```python
del dataframe["a"]
dataframe.pop("f")
dataframe
```
