```python
import  pandas as pd
```

```python
# series one dimensional data structure
list =["jawad","abbas","waqas","ajhdka","jdslajfj"]
series = pd.Series(list)
print(series)
print("type ",type(series))
# changing the values of index
series = pd.Series(list,index = ["a","s","d","f","g"])
print("\n")
print(" here i change the index values :\n",series)
```

```python
# slicing
# printing the element at index 3
print(series["d"])
print(series["f"])
```

```python
# dictionary
dict = {"language":["python","c","cpp","java"],"populairty":[100,230,43,54],"ranking":[1,3,4,2]}
ser = pd.Series(dict)
print(ser)
```

```python
series2 = pd.Series(32,index=[1,2,3,4])
series3 = pd.Series(12,index=[1,2,3,4,5,6,7])
print("adding both : \n",series2 + series3)
# if we doing this addition in numpy the numpy laibrary gives us a brodcasting error
# so here we also work with missing data
```

## DataFrame
it is a two dimensional data structure and
length of the array must be same

```python
list = [1,2,3,4,5,6,5,4,3,2,1]
var = pd.DataFrame(list)
print(var)
```

```python
dictionary = {"a":["addf","dfaf","fdred"],"b":["dfar","trer","erer"]}
varr = pd.DataFrame(dictionary,columns = ["b"])
# it only show the b columns of the dataframe
print("data frame : \n",varr)
print("\n")
var11 = pd.DataFrame(dictionary)
print("data frame : \n",var11)


```

```python
dict = {"a":[1,2,3,4,5],"b":[4,3,2,1,1],"c":[7,4,3,2,2],"d":[5,8,9,5,4]}
dictt = pd.DataFrame(dict)
print(dictt)
print("only c columns are here : \n",dictt["c"],[3])
```

### creating dataframe from series

### Arithematic operations



### addition

```python
data = {"a":[1,2,3,4,5],"b":[5,4,3,2,1]}
data_var = pd.DataFrame(data)
data_var
# addition
data_var["add"]= data_var["a"] + data_var["b"]
data_var 
```

### subtraction and multiplication

```python
data_var["sub"] = data_var["a"] - data_var["b"]
data_var
# multiplication
data_var["multiply"] = data_var["a"] * data_var["b"]
data_var
```

### average

```python
data_var["average"] = data_var["a"] + data_var["b"] /2
data_var
```

### conditonal operation

```python
new_data = {"a":[34,54,54,32,12,67],"b":[87,65,43,23,66,67]}
x = pd.DataFrame(new_data)
x
x["less_than_20 in column a"] = x["a"] <= 20
x
x["less_than_20 in column b"] = x ["b"] <= 50

x
```
