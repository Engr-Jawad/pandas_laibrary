# merging

```python
import numpy as np
import pandas as pd 
```

```python
var_1 = pd.DataFrame({"a":[1,2,3,5],"b":[5,6,7,8]})
var_2 = pd.DataFrame({"a":[1,2,4,5],"d":[5,4,3,2]})
pd.merge(var_1,var_2,on="a")
 # how{‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’, ‘left_anti’, ‘right_anti’},
```

```python
var_3 = pd.DataFrame({"a":[1,2,3,5],"b":[5,6,7,8]})
var_4 = pd.DataFrame({"a":[1,2,4,5],"b":[5,4,3,2]})
pd.merge(var_3,var_4,left_index=True,right_index=True,suffixes=("jawad","ali"))
```

# concat

```python
ser_one = pd.Series([1,2,3,4,5])
ser_two = pd.Series([5,4,3,2,1])
pd.concat([ser_one,ser_two])
```

```python
d1 = pd.DataFrame({"a":[1,2,3,5],"b":[5,6,7,8]})
d2= pd.DataFrame({"a":[1,2,4,5],"b":[5,4,3,2]})
pd.concat([d1,d2])
```

```python
d1 = pd.DataFrame({"a":[1,2,3,5],"b":[5,6,7,8]})
d2= pd.DataFrame({"a":[1,2,4,5],"b":[5,4,3,2]})
pd.concat([d1,d2],axis=1)
```

```python
d1 = pd.DataFrame({"a":[1,2,3,5],"b":[5,6,8,4]})
d2= pd.DataFrame({"a":[1,2,4],"b":[5,4,3]})
pd.concat([d1,d2],axis=1,keys=["d1"," d2"])
```

```python
d1 = pd.DataFrame({"a":[1,2,3,5]})
d2= pd.DataFrame({"a":[1,2,4],"b":[5,4,3]})
pd.concat([d1,d2],axis=1)
```
