# Replaced and interpolate

```python
import pandas as pd
```

```python
file_var = pd.read_csv("/home/jawad/Downloads/employees.csv")

file_var
```

```python
# to replace is ues to what replace with what value
file_var.replace(to_replace="Alice Smith",value="jawad ali")
# alice smith are replaced with jawad ali
```

```python
# years of experincing are replaced with 10
file_var.replace([0,1,2,3,4,5,6],10)
```

```python
# alphabatic data in the file are replaced with the help of regex keyword
file_var.replace("[a-zA-Z]","PAKISTAN",regex=True)
```

```python
# REPLACING DATA OF THE SPECFIC COLUMNS WITH THE HELP OF DICTIONARY
file_var.replace({"Name":"[A-Z]"},500,regex=True)
```
