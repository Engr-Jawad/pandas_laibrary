###  Problem: Student Exam Scores Analysis
You are given data for 5 students with their exam scores in three subjects. Perform basic data manipulation and analysis using pandas.

```python
import pandas as pd
import numpy as np
```

```python
dict = {"student":["ali","sara","john","emma","raj"],"math":[85,92,67,95,74],"science":[90,88,72,93,68],"english":[78,84,70,89,81]}
dfr = pd.DataFrame(dict)
print(dfr)
# dfr["average"] = (dfr["math"] + dfr["science"] + dfr["english"])/3
dfr["average"]= dfr[["math","science","english"]].mean(axis=1)
dfr
dfr["result"] = np.where(dfr["average"] >= 70,"pass","fail")
dfr



```
