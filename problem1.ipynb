### Problem 4 — Column Statistics

Create this DataFrame:

{
    "Math": [80, 90, 70, 60],
    "Physics": [75, 85, 95, 65]
}

Find:

mean
maximum value
minimum value

for each column.

```python
dict = {
    "Math": [80, 90, 70, 60],
    "Physics": [75, 85, 95, 65]
}
dataframee = pd.DataFrame(dict)
print(dataframee)
mean = np.mean(dataframee[1:3],axis=0)
print("mean of the columms : \n",mean)
print("maximum value in columns:\n",np.max(dataframee,axis=0))
print("minimum value in columns:\n",np.min(dataframee,axis=0))



print("_________________________________________________")
print("PANDAS STYLE ")
print("_________________________________________________")
print(dataframee.mean(axis=0))
print(dataframee.max(axis=0))
print(dataframee.min(axis=0))
```

### You are given a small dataset of students and their scores. Perform the following tasks to practice adding and removing data.



```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [88, 81, 95]
})
# adding new column
df.insert(3,"English",[90,84,88])


# calculating sum of the subjects
df.insert(4,"total",df[['Math','English']].sum(axis=1))

# adding a new row
new_row = pd.DataFrame([['Diana', 89, 93, 91,89+93+91]], columns=df.columns)
df = pd.concat([df, new_row], ignore_index=True)
print("before deletion data frame : \n\n\n",df)

# deleting the science columns
df.pop("Science")


# deleting the bob rows here we use inplace = true it modifies the orignal dataframe instead of returning a new dataframe
df.drop(index=1,inplace=True)


# reset index
df.reset_index(drop=True,inplace=True)
print('n\n\n\n')
print("Original DataFrame:")
print(df)
```
