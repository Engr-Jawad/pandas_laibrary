# Handling missing data (dropna abd fillna) in python pandas

```python
import pandas as pd
```

```python
csv_1 = pd.read_csv("/home/jawad/Downloads/employees.csv")
csv_1
```

# droping nan values from the file

```python
csv_1.dropna() # by default it drop only missing value rows
csv_1.dropna(axis=1) # it drop only the missing value columns
csv_1.dropna(axis=0) # it drop only the missing value rows
csv_1.dropna(how="any") # it drop the entire row in which all the values are nan 
```

```python
# csv_1
csv_1.dropna(subset=["Department"]) # droping the data in a specific column just like here i drop the nan values of department columns only
# csv_1.dropna(subset=["Name"])  here is no nan value in the name columns 
```

```python
csv_1.dropna(thresh= 2)
```

# filling missing data 

```python
csv_1.fillna(1111.1111)
```

# filling data within  a specfic columns

```python
csv_1.fillna({"Department":"computer science","Salary":0.00000})
```

# filling nan values like forward and backword values in a columns

```python
csv_1.ffill() # forward filling nan value filled with respect to the above the fill value
csv_1.bfill() # backword filling nan value filled with respect to the below the fill value
```
