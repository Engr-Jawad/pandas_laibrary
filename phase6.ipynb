# melt functions
<P> it is used to reshape the dataframe 

```python
import pandas as pd
```

```python
var = pd.DataFrame({"day":[1,2,3,4,5,6],"eng":[11,12,13,14,15,16],"math":[13,12,11,15,14,11]})
print(var)
print("------------------------")
# reshaping the dataframe
pd.melt(var,id_vars=["day"],var_name="class",value_name=("marks"))
```

# pivot function
<P> it is used to arrange the data according to specfic columns

```python
var = pd.DataFrame({"day":[1,2,3,4,5,6],"st_name":["a","b","c","d","e","f"],"eng":[11,12,13,14,15,16],"math":[13,12,11,15,14,11]})
print(var)
pd.pivot(var,index=["day"],columns="st_name")
piv = pd.pivot_table(var,index=["day"],columns="st_name",aggfunc="mean")
print(piv)


```
