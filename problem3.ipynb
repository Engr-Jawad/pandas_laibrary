## Medium Level Pandas Problem — Handling Missing Data

You are given sales data with missing values.

```python
import pandas as pd
import numpy as np
```

```python


data = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Printer"],
    "Price": [80000, np.nan, 3000, 25000, np.nan],
    "Quantity": [5, 10, np.nan, 3, 8],
    "Rating": [4.5, np.nan, 4.2, 4.8, np.nan]
}

df = pd.DataFrame(data)
print(df)
print("\n\n")

# Find total missing values in each column
print(df.isnull().sum())
print("\n\n")
# Fill missing Price values with the mean price
average = df["Price"].mean()
df["Price"] = df["Price"].fillna(average)
print("\n\n\n")
# Fill missing Quantity using forward fill
df["Quantity"] = df["Quantity"].ffill()
#  Fill Rating NaN with 0
df["Rating"] = df["Rating"].fillna(0)
# Create a new column called "Total_Value"
df["total_value"] = df["Price"] * df["Quantity"]
print(df)
# . Find:
# Product with highest total value 
# #Average rating
# Total quantity sold
df.loc[df["Product"].idxmax()]
df["Rating"] = df["Rating"].mean()
df["Quantity"] = df["Quantity"].sum()
df

```

# Student Performance Analysis

```python


data = {
    "Name": ["Ali", "Sara", "John", "Ayesha", "David", "Hina"],
    "Math": [78, 90, 67, np.nan, 88, 95],
    "Physics": [85, 92, 70, 60, np.nan, 98],
    "Chemistry": [80, 89, 72, 65, 91, np.nan],
    "Attendance": [90, 95, 80, 70, 85, 98]
}

df = pd.DataFrame(data)
# replacing missing value in math columns with the columns mean
df["Math"] = df["Math"].fillna(df["Math"].mean())
# replacing missing value in physics columns with the columns median
df["Physics"] = df["Physics"].fillna(df["Physics"].median())
# replacing missing value in chemistry columns with zero
df["Chemistry"] = df["Chemistry"].fillna(0)
# average of students
df["average"] = (df["Math"] + df["Chemistry"] + df["Physics"])/3
df["status"] = np.where(df["average"] >= 75 ,"pass","fail")
# show only student whose attandence is greater than 80
filter_student = (df["Attendance"] > 85 ) & (df["status"]=="pass")
print(filter_student)
print("\n\n")
# sort student by average in descending order
average_des = df.sort_values(by="average",ascending= False)
print(average_des)
print("-----------------------------------------------------")

# highest average
heighest_avg = df["average"].max()
print("highest average : ",heighest_avg)
# lowest average
lowest_Avg = df["average"].min()
print("lowest average : ",lowest_Avg)
# mean of the attandence
mean_Att = df["Attendance"].mean()
print("mean of the attandence : ",mean_Att)
# Find the student with the highest Average using:
std_high_avg = df.loc[df["average"].idxmax()]
print(std_high_avg)
print("-------------------------------------------")
df


```
